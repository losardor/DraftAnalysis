```python
import draftAnal as DA
import os

DA.resetFolder("./Data/WORKING")
```

### Collect information from input folder

```python
Data = []

for i, directory in enumerate(next(os.walk('./Data/INPUT/'))[1]):
    Data.append(DA.author(directory.lower()))
```

### Clean the filenames and assign each author their draft

```python
for i, author in enumerate(Data):
    author.cleanfilenames()
    Data[i] = author
```

```python
#from firts workshop
print(Data[0].drafts[1].Cname)
#from second workshop
print(Data[4].drafts[1].Cname)
```

### Count the characters of each draft and remove those for which the analysis is impossible

```python
for i, author in enumerate(Data):
    for j,draft in enumerate(author.drafts):
        draft.CharacterCount()
        if draft.CharCount < 600:
            print(draft.Cname)
        else:
            author.drafts[j] = draft
    Data[i] = author
```

```python
for i, author in enumerate(Data):
    author.createContigs()
    Data[i] = author
```

```python
for i, author in enumerate(Data):
    author.FindLastMacro(verbose=True)
    Data[i]=author
```

```python
from pprint import pprint
from scipy.signal import find_peaks
import matplotlib.pyplot as plt


for author in Data:
    print(author.name)
    X = author.drafts[author.Last4].contigData.reset_index()["distance"]
    peaks, _ = find_peaks(X, threshold=0.2)
    plt.plot(X)
    plt.plot(peaks, X[peaks], "x")
    plt.show()
    if peaks.size>0:
        pp = author.drafts[author.Last4].contigData.iloc[peaks,4]
        tagtoind = {draft.tag:i for i, draft in enumerate(author.drafts)}
        pktime = pp.reset_index().datetime.dt.strftime("%d_%m_%H_%M").values[0]
        print(pktime)
```
